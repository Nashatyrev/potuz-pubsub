```kotlin
import org.jetbrains.kotlinx.dataframe.api.*

val param1Col = columnOf(1, 1, 1, 1, 1, 2, 2, 2, 2).named("param1")
val param2Col = columnOf(10, 10, 10, 15, 15, 10, 10, 10, 10).named("param2")
val timeCol = columnOf(0.0, 0.1, 0.5, 0.8, 1.5, 0.2, 0.5, 0.8, 1.0).named("time")
val valueCol = columnOf(0, 1, 2, 3, 4, 2, 3, 4, 5).named("value")
val value1Col = columnOf(2, 3, 4, 5, 3, 4, 5, 1, 1).named("value1")

val df = dataFrameOf(param1Col, param2Col, timeCol, valueCol, value1Col)
val df1 = df.groupBy { param1Col and param2Col }.updateGroups { remove { param1Col and param2Col } }.into("group")
df1
```

```kotlin
val df2 = df1.explode { group }.flatten { group }
df2
```

```kotlin
val df3 = df2.merge { param1 and param2 }.into("param")

plot(df3) {
    line{
        x(time)
        y(value)
        color("param")

        points {
            x(time)
            y(value)
            size = 3.5
            symbol = Symbol.CIRCLE_OPEN
            color("param")
        }
    }
    
    line{
        x(time)
        y(value1)
        color("param")
        type = LineType.DASHED
    }
}
```
