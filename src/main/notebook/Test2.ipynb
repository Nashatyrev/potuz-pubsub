```kotlin
import net.nashat.*
import net.nashat.ChunkSelectionStrategy.*

val resDf000 =
    PotuzSimulation.runAll(
        (10L..20L).flatMap { randomSeed ->
            listOf(3, 4, 5)
                .map { peerCount ->
                    SimConfig(
                        peerCount = peerCount,
                        erasure = Erasure.NoErasure,
                        numberOfChunks = 20,
                        rsChunkSelectionStrategy = PreferRarest,
                        latencyRounds = 10,
                        randomSeed = randomSeed
                    )
                }
        }
    )
        .deriveExtraResultsExploded()
```

```kotlin
val qwqwq = resDf000
    .selectLastForEachGroup { config }
```

```kotlin
val random = kotlin.random.Random(42)
fun generateData(category: String) =
    dataFrameOf((0..<10).map { it.toString() }) { List(100) { random.nextDouble(0.0, 100.0) } }
        .gather(*(0..<10).map { it.toString() }.toTypedArray())
        .into("num", "value").add("category") { category }

val dataset = generateData("category0")
    .concat(generateData("category1"))
    .concat(generateData("category2"))

dataset.groupBy("category").plot {
    boxplot("num", "value") {
        boxes {
            position = Position.dodge()
        }
    }
}
```
