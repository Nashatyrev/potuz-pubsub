```kotlin
%use dataframe, kandy(0.8.0-dev-59)
```

```kotlin
import net.nashat.*

val resDf = PotuzIO().readResultsFromJson("../../../results/result.json")
    .normalizePotuzLoadedResults()
    .deriveExtraResults()
```

```kotlin
// Print configs overview
resDf
    .select { config }
    .flatten { all() }
    .gather { all() }.into("Param", "Value")
    .distinct()
    .groupBy { "Param"() }.values()
```

```kotlin
import net.nashat.Erasure
import net.nashat.myPlotGroupDeliveredPartsAndMessageTypeCounts

// NoErasure doesn't make much sense with peerCount > 6
fun capPeerCountForErasure(peerCount: Int, erasure: Erasure) =
    if (erasure in setOf(Erasure.NoErasure, Erasure.NoErasureOneMesh)) 6 else peerCount

resDf
    .filter {
        config.numberOfChunks == 40
                && config.erasure in setOf(Erasure.NoErasure, Erasure.RsX2, Erasure.RsX3, Erasure.RLNC)
                && config.peerCount == capPeerCountForErasure(40, config.erasure)
    }
    .explode { result }
    .myPlotGroupDeliveredPartsAndMessageTypeCounts(adjustX2 = -60) {
        config.erasure
    }
```

```kotlin
resDf
    .filter {
        config.numberOfChunks == 10
                && config.erasure == Erasure.NoErasure
                && config.peerCount in setOf(6, 10, 100)
    }
    .explode { result }
    .myPlotGroupDeliveredPartsAndMessageTypeCounts(adjustX2 = -60) { 
        config.peerCount 
    }
```

```kotlin
val oneChunkDf = resDf
    .filter {
        config.numberOfChunks == 1
                && config.erasure == Erasure.NoErasure
                && config.peerCount in setOf(6, 10, 100)
    }
//    .removeNonChangingConfigColumns()
    .flatten { config.allCols() }
    .convert { config.peerCount }.toStr()
    .explode { result }
```

```kotlin
import net.nashat.myPlotLinesWithEndingVLine

oneChunkDf.myPlotLinesWithEndingVLine(
    xCol = { result.derived.relativeRound },
    yCol = { result.derived.doneMsgFraction },
    groupCol = { config.peerCount }) {

    layout {
        size = 1524 to 600
    }
}
```

```kotlin
val noEcDf = resDf
    .filter {
        config.numberOfChunks in setOf(1, 2, 4, 10, 20, 100)
                && config.erasure == Erasure.NoErasure
                && config.peerCount == 6
    }
    .flatten { config.allCols() }
    .explode { result }
```

```kotlin
noEcDf.myPlotGroupDeliveredPartsAndMessageTypeCounts { config.numberOfChunks }
```
