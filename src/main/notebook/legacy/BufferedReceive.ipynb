```kotlin
import net.nashat.*
import net.nashat.ChunkSelectionStrategy.*

val cfgs =
    listOf(1, 2, 4, 20, 40)
        .map { receiveBufferSize ->
            PotuzSimulationConfig(
                params = PotuzParams(
                    numberOfChunks = 40,
                    messageBufferSize = receiveBufferSize,
                    chunkSelectionStrategy = PreferLater,
                    rsParams = RSParams(2, true)
//                    rlncParams = RLNCParams()
                ),
                peerCount = 20,
                isGodStopMode = true,
                randomSeed = 1
            )
        }

val res = PotuzSimulation.runAll(cfgs)

val resDf = res
    .normalizePotuzLoadedResults()
    .deriveExtraResults()
// Print configs overview
resDf
    .select { config }
    .flatten { all() }
    .gather { all() }.into("Param", "Value")
    .distinct()
    .groupBy { "Param"() }.values()
```

```kotlin
resDf
    .explode { result }
    .myPlotGroupDeliveredPartsAndMessageTypeCounts(adjustX2 = 20) {
        config.messageBufferSize
    }
```
